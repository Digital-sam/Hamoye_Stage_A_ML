### Question 5
#### first we import the libaries and dependencies

```python
from scipy.stats import skew
from scipy.stats import kurtosis
```

```python
F_qty_burned = df['fuel_qty_burned']
kurt = kurtosis(F_qty_burned)
sk = skew(F_qty_burned)

print('Kurtosis: ' + str(kurt))
print('skew: ' + str(sk))
```
